# 🧮 Fórmula do IMC

O Índice de Massa Corporal será calculado utilizando a seguinte fórmula:

\[
IMC = \frac{peso}{altura^2}
\]

---

# 📊 Classificação do IMC

| IMC | Classificação |
|:---:|:--------------|
| Menor que 18,5 | Abaixo do peso |
| De 18,5 até 24,9 | Peso normal |
| De 25,0 até 29,9 | Sobrepeso |
| Igual ou maior que 30 | Obesidade |

```python
# Exibe o título
print("#" * 20, "BEM-VINDO À NARA FIT", "#" * 20)

# Lista dos alunos cadastrados
alunos = []

#sistema funciona até o usuário decidir sair
while True:
    print("\n1. Cadastrar aluno")
    print("2 - Listar alunos cadastrados")
    print("3 - Sair")

    try:
        # transforma a opção digitada em inteiro
        opcao = int(input("Digite a opção desejada: "))
    except ValueError:
        print("Opção inválida. Por favor, digite um número.")
        continue
    if opcao == 1:
        print("\nCadastro de Aluno")
        nome = input("Digite o nome do aluno: ").strip()
        if nome == "":
            print("Nome inválido. Por favor, digite um nome válido.")
            continue

        try: # 3º Recebe idade, peso e altura (Etapa 4)

            idade = int(input("Digite a idade do aluno: "))
            peso = float(input("Digite o peso do aluno (em kg): ").replace(",", "."))
            altura = float(input("Digite a altura do aluno (em metros): ").replace(",", "."))

        except ValueError:
            print("Idade, peso ou altura inválidos. Por favor, digite valores numéricos válidos.")
            continue

        if idade <= 0 or peso <= 0 or altura <= 0:
            print("Idade, peso ou altura inválidos. Por favor, digite valores positivos.")
            continue
       
        imc = peso / (altura ** 2)  # Calcula o IMC

        if imc < 18.5:
            classificacao = "Abaixo do peso"
        elif imc < 25:
            classificacao = "Peso normal"
        elif imc < 30:
            classificacao = "Sobrepeso"
        else:
            classificacao = "Obesidade"

            #Cria um dicionário com os dados do aluno
        aluno = {
            "nome": nome,
            "idade": idade,
            "peso": peso,
            "altura": altura,
            "imc": imc,
            "classificacao": classificacao
        }
        alunos.append(aluno) #salva o dicionário na lista de alunos
        print(f"\nAluno {nome} cadastrado com sucesso!")
        print(f'Nome: {nome}, Idade: {idade}, Peso: {peso} kg, Altura: {altura} m, IMC: {imc:.2f}, Classificação: {classificacao}')

    elif opcao == 2:
        print("\nAlunos Cadastrados:")
        if len(alunos) == 0:
            print("Nenhum aluno cadastrado.")
        else:
            for aluno in alunos:
                print(f'Nome: {aluno["nome"]}, Idade: {aluno["idade"]}, Peso: {aluno["peso"]} kg, Altura: {aluno["altura"]} m, IMC: {aluno["imc"]:.2f}, Classificação: {aluno["classificacao"]}')
    elif opcao == 3:
        print("\nSaindo do programa...")
        print(f"Total de alunos cadastrados: {len(alunos)}")
        break
```
